```python
# Dependencies
import pandas as pd
import numpy as np
import time
import urllib.request
from bs4 import BeautifulSoup as bs
import pymongo
from splinter import Browser
```

```python
start = time.time()
# visit NASA website
executable_path = {'executable_path': '/usr/local/bin/chromedriver'}
browser = Browser('chrome', **executable_path, headless = False)

mars_url = "https://mars.nasa.gov/news/"
browser.visit(mars_url)

content = urllib.request.urlopen(mars_url).read()
print(content)
```

```python
mars_html = browser.html
```

```python
soup = bs(mars_html, "lxml")
print(soup.prettify())
```

```python
news_title = soup.find('div',class_="bottom_gradient").text
print(news_title)
```

```python
news_ptab = soup.find('div', class_="rollover_description_inner").text
print(news_ptab)
```

```python
# visit JPL website
jpl_url = "https://www.jpl.nasa.gov/spaceimages/?search=&category=Mars"
browser.visit(jpl_url)

content = urllib.request.urlopen(jpl_url).read()
print(content)
```

```python
jpl_html = browser.html
```

```python
soupJPL = bs(jpl_html, 'lxml')
print(soupJPL.prettify())
```

```python
browser.click_link_by_partial_text("FULL IMAGE")
browser.click_link_by_partial_text("more info")
```

```python
image_html = browser.html

soupJPLimg = bs(image_html, "lxml").findAll("a")
print(soupJPLimg)
```

```python
JPL_image_url = "https://www.jpl.nasa.gov/" + str(soupJPLimg)
print(JPL_image_url)
```
